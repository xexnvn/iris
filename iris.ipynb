```python
import seaborn as sns
%config InlineBackend.figure_format = "svg"
```

```python
df = sns.load_dataset("iris")
```

```python
df.info()
```

```python
sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"]
            ).set(title = "Eva: Sepal length vs Sepal width", 
                  xlabel = "Sepal length", 
                  ylabel = "Sepal length")
```

```python
sns.regplot(x = df["petal_length"],
            y = df["petal_width"],
            scatter_kws = {"color": "red"}
            ).set(title = "Eva: Petal length vs Petal width",
                  xlabel = "Petal length",
                  ylabel = "Petal width")
```

```python
sns.set_style("white")
sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"]
            ).set(title = "Eva: Sepal length vs Sepal width, White style", 
                  xlabel = "Sepal length", 
                  ylabel = "Sepal length")
```
