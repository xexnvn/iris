# Iris Analysis

## Import the Required Libraries

```python
import seaborn as sns
%config InlineBackend.figure_format = "svg"
```

```python
import ssl
ssl._create_default_https_context = ssl._create_stdlib_context
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the Dataset

```python
df = sns.load_dataset("iris")
```

```python
df.info()
```

## Regplot

```python
sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"]
            ).set(title = "Eva: Sepal length vs Sepal width", 
                  xlabel = "Sepal length", 
                  ylabel = "Sepal length")
```

```python
sns.regplot(x = df["petal_length"],
            y = df["petal_width"],
            scatter_kws = {"color": "red"}
            ).set(title = "Eva: Petal length vs Petal width",
                  xlabel = "Petal length",
                  ylabel = "Petal width")
```

```python
sns.set_style("white")
sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"]
            ).set(title = "Eva: Sepal length vs Sepal width, White style", 
                  xlabel = "Sepal length", 
                  ylabel = "Sepal length")
```

# Bubble Plot

```python
gapminder = pd.read_csv("https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv")
```

## Inspect Data

```python
gapminder.info()
```

```python
gapminder.head()
```

```python
gapminder.tail()
```

```python
gapminder.continent.unique()
gapminder.year.unique()

data = gapminder.loc[gapminder.year == 2007]
```

```python
data.shape
```

```python
sns.set_theme(style = "whitegrid")
sns.scatterplot(data = data,
                x = "gdpPercap",
                y = "lifeExp",
                size = "pop",
                hue = "country",
                legend = False,
                sizes = (20, 2000)
                ).set(title = "Eva: GDP per Capita vs Life expectancy", 
                      xlabel = "GDP per Capita", 
                      ylabel = "Life expectancy")
plt.show()
```

```python
data = gapminder.loc[(gapminder.year == 2007) & (gapminder.continent == "Asia")]
```

```python
sns.set_theme(style = "whitegrid")
sns.scatterplot(data = data,
                x = "gdpPercap",
                y = "lifeExp",
                size = "pop",
                hue = "country",
                legend = False,
                sizes = (20, 2000)
                ).set(title = "Eva: GDP per Capita vs Life expectancy, Asia", 
                      xlabel = "GDP per Capita", 
                      ylabel = "Life expectancy")
plt.show()
```

```python
data = gapminder.loc[(gapminder.year == 2007) & (gapminder.continent == "Europe")]
data.shape
```

```python
sns.set_theme(style = "whitegrid")
sns.scatterplot(data = data,
                x = "gdpPercap",
                y = "lifeExp",
                size = "pop",
                hue = "country",
                legend = False,
                sizes = (20, 2000)
                ).set(title = "Eva: GDP per Capita vs Life expectancy, Europe", 
                      xlabel = "GDP per Capita", 
                      ylabel = "Life expectancy")
plt.show()
```
